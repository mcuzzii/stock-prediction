```python
import numpy as np
from pdf2image import convert_from_path
from paddleocr import PPStructure

# 1. Initialize the Table Engine
table_engine = PPStructure(layout=True, table=True, show_log=False)

pdf_path = 'data/raw/news/test.pdf'

# 2. Convert PDF pages to PIL Image objects
# 'dpi=300' is the sweet spot for OCR accuracy
pages = convert_from_path(pdf_path, dpi=300)

for i, page in enumerate(pages):
    # 3. Convert PIL image to OpenCV/NumPy format (BGR)
    # PaddleOCR expects a BGR numpy array
    img_array = np.array(page)
    img_array = img_array[:, :, ::-1] # Convert RGB to BGR

    # 4. Run Table Extraction
    result = table_engine(img_array)
    
    print(f"Processed Page {i+1}")
    
    # Do something with the result (e.g., access the HTML of the table)
    for region in result:
        if region['type'] == 'table':
            print(f"Table found: {region['res']['html'][:100]}...")
```
